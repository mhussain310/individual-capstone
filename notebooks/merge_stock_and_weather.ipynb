```python
import pandas as pd
```

```python
stock_df = pd.read_csv('../data/raw/hourly_stock_data.csv')
df_hourly = pd.read_csv('../data/test/hourly_historical_weather.csv')
```

```python
df_hourly["time"] = pd.to_datetime(df_hourly["time"])

df_hourly["time"] = df_hourly["time"].dt.strftime("%Y-%m-%d %H:%M:%S")

df_hourly["time"] = df_hourly["time"].astype(str)

df_hourly
```

```python
# Convert 'timestamp' column to datetime format
stock_df['timestamp'] = pd.to_datetime(stock_df['timestamp'])

# Round to the nearest hour
stock_df['timestamp'] = stock_df['timestamp'].dt.round('h')


# Drop duplicates, keeping the first occurrence
stock_df.drop_duplicates(subset=['timestamp'], keep='first', inplace=True, ignore_index=True)

stock_df['timestamp'] = stock_df['timestamp'].astype(str)

stock_df
```

```python
# Merge the DataFrames on 'time' and 'timestamp'
merged_df = pd.merge(df_hourly, stock_df, left_on='time', right_on='timestamp', how='inner')

# Drop one of the duplicate datetime columns if needed
merged_df.drop(columns=['timestamp'], inplace=True)

merged_df
```

```python
def format_date(df, column="time"):
    # Convert 'time' column to datetime format
    df[column] = pd.to_datetime(df[column])

    # Format it to 'YYYY-MM-DD HH:MM:SS'
    df[column] = df[column].dt.strftime("%Y-%m-%d %H:%M:%S")

    # Convert back to object(string)
    df[column] = df[column].astype(str)

    return df


def floor_date_to_hour(df, column="timestamp"):
    # Convert 'timestamp' column to datetime format
    df[column] = pd.to_datetime(df[column])

    # Floor to nearest hour
    df[column] = df[column].dt.floor("H")

    # Drop duplicates, keeping the first occurrence
    df.drop_duplicates(subset=[column], keep="first", inplace=True, ignore_index=True)

    # Convert back to object(string)
    df[column] = df[column].astype(str)

    return df

def merge_current_stock_and_weather(weather, stock, file_name):
    formatted_weather = floor_date_to_hour(
        format_date(weather, column="local_time"), column="local_time"
    )
    formatted_stock = floor_date_to_hour(stock)

    # Merge the DataFrames on local_time and timestamp
    merged_df = pd.merge(
        formatted_weather,
        formatted_stock,
        left_on="local_time",
        right_on="timestamp",
        how="inner",
    )

    # Drop one of the duplicate datetime columns
    # merged_df.drop(columns=["timestamp"], inplace=True)

    return merged_df


merged_current_data = merge_current_stock_and_weather(
        current_weather, current_stock, "merged_current_data.csv"
    )

merged_current_data
```
